```python
#Import Packages
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
import datetime
from copy import deepcopy

#Define headers for scraping
headers = {'User-Agent': 
           'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.106 Safari/537.36'}
```

```python
page = "https://www.transfermarkt.com/premier-league/spieltagtabelle/wettbewerb/IR1/saison_id/2022/spieltag/1"
pageTree = requests.get(page, headers=headers)
pageSoup = BeautifulSoup(pageTree.content, 'html.parser')
```

```python
seasons = pageSoup.find("div", class_ = "inline-select").text.strip().split("\n")
```

```python
md_str = pageSoup.find("select", {"name":"spieltag"}).text.strip().split('\n')

matchdays = []

for i in range(len(md_str)):
    matchdays.append(''.join(ch for ch in md_str[i] if ch.isdigit()))
```

```python
#Look for all of the tables on the page
print('Classes of each table:')
for table in pageSoup.find_all('table'):
    print(table.get('class'))
```

```python
# Creating list with all tables
tables = pageSoup.find_all('table')

#Extract the one table we want
table = pageSoup.find('table', class_='items')

#Extract our desired column headers
rows = table.find("tr").find_all("th")

cols = []

for row in rows:
    if row.text.strip() != '':
        cols.append(row.text.strip())
    
    else:
        cols.append('Matches')
        
table_rows = list(table.find("tbody").find_all("tr"))
data = []

for i in range(len(table_rows)):
    dummy_list = []
    
    for table_row in table_rows:
        dummy_list.append(table_row.find_all("td")[i].text.strip())
    
    data.append(dummy_list)

data.pop(1) 
```

```python
df = pd.DataFrame(data).transpose()
df.columns=cols
df
```
